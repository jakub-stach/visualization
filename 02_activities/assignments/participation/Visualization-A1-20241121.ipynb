```python
#importing the necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
```

```python
#setting random seed and creating 2 sets of Y variables

np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint (0,100,50)
```

```python
#making a plot line with both sets of Y variables

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
#practicing adding a legend

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "person 1")
ax.plot(x,y2, label = "person 2")
ax.legend(loc='upper right') # you can specify the location of the legend
fig.show()
```

```python
#modifying the legend

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y1, label = "person 1")
ax.plot(x,y2, label = "person 2")
ax.legend(loc='lower right',
          frameon = True, # adds a frame arround a legend
          fontsize = 12, #specifies font size
          ncol = 2, #specify number of columns
          shadow = True #adds a shadow to the label box)
#fig.show()
```

```python

#specifying the location of the the legend outside the chart using bbox_to_anchor
fig, ax  = plt.subplots(figsize=(5,3))
ax.plot(x,y1,label="person 1")
ax.plot(x,y2, label="person 2")
ax.legend(loc='upper left',
          bbox_to_anchor = (1,0.5)) # the units are relating to the figure where the 1 refers to 100% and goes down into decimals
fig.show()
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="person 1") #plotting a scatter this time
ax.scatter(x,y2, label="person 2")
ax.legend(loc='lower right')
ax.text(10,95, "this value is important!")
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.scatter(x,y1,label="person 1") #plotting a scatter this time
ax.scatter(x,y2, label="person 2")
ax.legend(loc='lower right')
#adding a particular comment in regards to a specific value
ax.text(10,95, "this value is important!", 
        #specifying different stylistic things for the annotation
        ha='center', #alignment of text
        color = 'red', #font colour
        size = 20) #font size
```

```python
fig, ax = plt.subplots()
ax.axis([0,10,0,10]) #creating a  simple set of empty axes

#positioning text annotations to specicic points
ax.text(1,5, ", Data: (1,5)", transform=ax.transData) 
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "person 1")
ax.scatter(x,y2, label = "person 2")
ax.legend(loc = "lower right")

#adding an arrow to point out what's crucial
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(facecolor= 'black', shrink = 0.05)) #specifying information as to the arrow
```

```python
# same as above but making more modifications to the arrow
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y1, label = "person 1")
ax.scatter(x,y2, label = "person 2")
ax.legend(loc = "lower right")

#adding an arrow to point out what's crucial
ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94),
            arrowprops=dict(arrowstyle = 'wedge',
                            color = 'hotpink')) #more details on the arrow
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'person 1')
ax.scatter(x,y2, label = "person 2")
ax.legend(loc='lower right')
#removing tick marks and labels on our axis
ax.yaxis.set_major_locator(plt.NullLocator()) #removes ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) #removes labels only
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'person 1')
ax.scatter(x,y2, label = "person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MaxNLocator(3))  #defines maximum amount of ticks we want on an axis, in this case its 3 tickmarks
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'person 1')
ax.scatter(x,y2, label = "person 2")
ax.legend(loc='lower right')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #this sets tickmarks at a particular interval, in this case every 5
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'person 1')
ax.scatter(x,y2, label = "person 2")
ax.legend(loc='lower right')

plt.xticks(rotation=45, ha='right') #rotates axis labels and horizonatl alignment
```

```python
font1 = {'family': 'serif', 'color': 'indigo'}

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y1, label = 'person 1')
ax.scatter(x,y2, label = "person 2")
ax.legend(loc='lower right')
#modifying our x axis title to add in more details
plt.xlabel('shine new x axis!', fontsize = 18, fontdict = font1)
```
