```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
from PIL import Image
import requests
from io import BytesIO
```

```python
x = np.array(["luffy","zoro","nami","usopp","sanji"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7,3))
#layeering 2 types of vis on the same object
ax.bar(x,y1, color = "indigo")
ax.plot(x,y2,color="red")
```

```python
#calculating the standard deviation of our data

y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color="red")

#add the SD using the error bar function
ax.errorbar(x, #our x values
            y2,#our y values
            yerr = y2_sd, # yerr = error for y
            fmt = "none", #format = fmt
            ecolor = "purple",
            elinewidth = 4,
            capsize = 6,
            capthick = 4, #just like other parts of our graphs, we can modify the SD display
            errorevery = 2) #we can also indicate the intervals for the SDs
```

```python
#adding pictures to our graphs

response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png') #gets file
image_file = BytesIO(response.content) #names it

image = Image.open(image_file) #shows it
```

```python
#plots a basic plot graph
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = "red")

#this adds in an image box
ax_image = fig.add_axes([0.1, #the x coordinate, on figure not axes
                         0.11, #the y coordinate, on figure not axxes
                         0.15, #image width
                         0.35]) #image height

#this adds the image we prepared before

ax_image.imshow(image)
ax_image.axis('off')
plt.show()
```
