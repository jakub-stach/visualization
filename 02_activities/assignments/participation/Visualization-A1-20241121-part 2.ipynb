```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#creating 2 sets of  new data
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["luffy", 'zoro', 'nami', 'usopp', 'sanju'])
y2 = np.array([110,180,240,99,220])
```

```python
#creates twi dufferebt

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
#plt.style.use('fivethirtyeight')

fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
#someaxes is a name for a group of axes
#the mosaic creates a tapestry or mozaic like strucutre, by giving the coordinates and orientation of the axes
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2', 'ax3']],
                                    figsize = (7,4),
                                    layout = 'constrained')
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('a big label',fontsize=18)
someaxes["ax2"].set_xlabel('another one', fontsize=18)
someaxes["ax3"].set_xlabel('they dont want to see us winning', fontsize='18')

# makes things messy unless we add a contrained layout
```
