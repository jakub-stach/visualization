```python
#importing our standard libraries

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset("tips") #seaborn comes with some sample data sets, one of these is tips
print(tips)
```

```python
sns.lineplot(data=tips, #choosing our data set
             x='total_bill', #defining our x variable
             y='tip') #defining our y variable
```

```python
#just like matplotlib, seaborn also comes with a set of styles

sns.set_style("whitegrid")

sns.lineplot(data=tips, #choosing our data set
             x='total_bill', #defining our x variable
             y='tip') #defining our y variable
```

```python
tipgraph = sns.lineplot(data=tips,
                        x='total_bill',
                        y='tip')
#just like matplotlib we can do the same things in terms of labeling the graph
tipgraph.set(title='tips vs total bill',
             xlabel='total bill($)',
             ylabel='tip amount($)')
```

```python
fig = plt.subplots(figsize=(10,3))
tipgraph = sns.lineplot(data=tips,
                        x = 'total_bill',
                        y = 'tip',
                        color = 'hotpink',
                        linestyle ='--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')

tipgraph.set(title = 'tips vs total bill',
             xlabel = 'total bill ($)',
             ylabel = 'tip amount ($)')
```

```python
#an example of seaborn having more functionality in terms of represneting different variables with different visuals

tipgraph = sns.scatterplot(data=tips,
                            x = "total_bill",
                            y = 'tip',
                            style = 'time', #change our style according to the time
                            hue = 'day',
                            palette = ['purple','hotpink','deepskyblue','yellowgreen'])

tipgraph.set(title = 'tips vs total bill',
             xlabel = 'total bill ($)',
             ylabel = 'tip amount ($)')
```

```python
#pairplot allows us to compare multidimensional relationships between variables

sns.pairplot(
    data=tips,
    hue = 'day'
)
```

```python
#relplot allows us to explor specific levels within variables

daysplot = sns.relplot(
    data=tips,
    x="total_bill",
    y='tip',
    hue='sex',
    col='day',
    kind='scatter',
    col_wrap =2
)
```
