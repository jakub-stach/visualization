```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613) # setting seed for repeatability
x = np.arange(50) 
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y) #scatter specifies to create a scatterplot
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('total growth over time') #specifies the title of the axes
ax.set_ylabel('total growth') #specifies the y title
ax.set_xlabel('years since start') #specifies the x title
fig.tight_layout()
```

```python
#creating a font dictionary that has font descriptions

font1 = {'family':'sans-serif', 'color':'blue', 'size': 20}
font2 = {'family':'monospace', 'color':'green', 'size': 14}
```

```python
font1 = {'family':'sans-serif', 'color':'blue', 'size': 20}
font2 = {'family':'monospace', 'color':'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)
ax.set_title('total growth over time', fontdict=font1) #identifies the font for the title
ax.set_ylabel('total growth', fontdict=font2) 
ax.set_xlabel('years since start',fontdict=font2) 
```

```python
fig, ax = plt.subplots(figsize=(5,3))
#changing how the plot is displayed/more customized

#the additional information like markers, linestyle, color, and font are all available from the python documentation
ax.plot(x,
        y,
        marker='*',
        color='indigo',
        linestyle = '--',
        linewidth = 2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))

#you can specify the colour with a hex code as well
ax.plot(x,y,marker='*', color = '#6425b9', linestyle = '--', linewidth = 2)
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,marker='*',
        markersize=12,
        color='#7426b9',
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,marker='*',
        markersize=12,
        color='#7426b9',
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')

ax.grid(axis = 'y') #adding a grid to it, if we specify what it is then we add a vertical axis

# you can even modify the colour and other appearance factors of the gridlines
```
